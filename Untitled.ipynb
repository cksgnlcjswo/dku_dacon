```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
import numpy as np
from pathlib import Path
import pandas as pd
import seaborn as sns
from tqdm.notebook import tqdm
import warnings
```

```python
rcParams['figure.figsize'] = (16, 8)
plt.style.use('fivethirtyeight')
pd.set_option('max_columns', 100)
pd.set_option("display.precision", 4)
warnings.simplefilter('ignore')
```

```python
data_dir = Path('../input/')
trn_file = data_dir / 'train.csv'
seed = 42
```

```python
trn = pd.read_csv(trn_file, index_col=0)
print(trn.shape)
trn.head()
```

```python
sns.pairplot(trn.sample(n=10000), vars=['u','dered_u'], hue='class',size=5)
```

```python
X = trn[trn['class']!=0][['u']].values
y = trn[trn['class']!=0]['dered_u'].values
print(X.shape, y.shape)
```

```python
rmse = lambda y,p: np.sqrt(np.mean((y-p)**2))
```

```python
alpha = .001 #running rate
theta0= .1
theta1= .1
epoch = 4 
```

```python
x = X.flatten()
print(x.shape)
```

```python
rmses = []
for i in range(epoch):
    for x_i, y_i in zip(x, y):
        p_i =  theta0 + theta1 * x_i
        theta0 = theta0 + alpha * (y_i - p_i)
        theta1 = theta1 + alpha * (y_i - p_i) * x_i
    rmses.append(rmse(y, theta0 + theta1 * x))
```

```python
theta1, theta0
```

```python
rmses
```

```python
plt.plot(rmses)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, theta0+theta1*x,color='orange',linewidth=.5)
plt.xlim(0,40)
plt.ylim(0,40)
plt.xlabel('u')
plt.ylabel('dered_u')
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
p=lr.predict(X)
```

```python
rmse(y,p)
```

```python
lr.coef_,lr.intercept_
```
